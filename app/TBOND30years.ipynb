```python
from xtquant import xtdata
import backtest as bt
from backtest import Context 
import pandas as pd
import datetime
```

```python
code_list = ['511130.SH']

period = 'tick'

fields=['time', 'lastPrice', 'open', 'close', 'high', 'low', 'volume', 'amount', 'preClose']

dividend_type = 'none'

start_date_str = '20250115'

end_date_str = '20250117'

count = -1

xtdata.download_history_data2(code_list, period, start_date_str, end_date_str)
    
ret = xtdata.get_market_data_ex(field_list=fields,
                                    stock_list=code_list,
                                    period=period,
                                    start_time=start_date_str,
                                    end_time=end_date_str,
                                    count=count,
                                    dividend_type=dividend_type
            )
```
